```python
msg = "喵"
print('哪部影集適合你？你想看哪種的？')
while (msg != "好期待"):
    msg = input(">>")
    if '亞洲的'  in msg:
        print ('那要驚悚的白痴的幽默的還是肥皂劇')
        msg = input(">>")
        if '白癡的' in msg:
            print('看這個!!')
            display (B99)
        msg = input(">>")
        if '幽默的' in msg:
            print('這個這個～')
            display(RM)
        msg = input(">>")
        if '驚悚的' in msg:
            print('看這個')
            display (GFNW)
    if '歐美的' in msg:
        print('那要白痴的幽默的還是肥皂劇')
        msg = input('>>')
        if '白癡的' in msg:
            print('看這個!!')
            display (B99)
        if '幽默的' in msg:
            print('這個這個～')
            display(RM)
    else :
        print ('超好看喔')
```
